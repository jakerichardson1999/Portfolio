```python
#function to change the values of the data from words the numbers
def change_values(value):
    if value == 'Agree a little':
        return 5
    elif value == 'Agree moderately':
        return 6
    elif value == 'Disagree moderately':
        return 2 
    elif value == 'Disagree a little':
        return 3 
    elif value == 'Neither agree nor disagree':
        return 4
    elif value == 'Agree strongly':
        return 7 
    elif value == 'Disagree strongly':
        return 1
    else:
        return value
```

```python
dfnew['ExtroScore'] = dfnew['Extroverted, enthusiastic'].apply(change_values)
```

```python
dfnew['CritScore'] = dfnew['Critical, quarrelsome\xa0'].apply(change_values)
dfnew['DepScore'] = dfnew['Dependable, self disciplined\xa0'].apply(change_values)
dfnew['AnxScore'] = dfnew['Anxious, easily upset'].apply(change_values)
dfnew['OpenScore'] = dfnew['Open to new experiences, complex'].apply(change_values)
dfnew['ResScore'] = dfnew['Reserved, quiet\xa0'].apply(change_values)
dfnew['SympScore'] = dfnew['Sympathetic, warm'].apply(change_values)
dfnew['DisScore'] = dfnew['Disorganised, careless'].apply(change_values)
dfnew['CalmScore'] = dfnew['Calm, emotionally stable'].apply(change_values)
dfnew['ConScore'] = dfnew['Conventional, uncreative'].apply(change_values)
```

```python
dfnew.head()
```

```python
#nwo apply formula to those scores to get final score for each trait
def calculate(a, b):
    return (7 - a + 1 + b)/ 2
```

```python
dfnew['Agreeableness'] = dfnew.apply(lambda row: calculate(row['CritScore'], row['SympScore']), axis = 1)
```

```python
dfnew.head()
```

```python
#looks good, now lets apply that formula to the other values
dfnew['Extraversion'] = dfnew.apply(lambda row: calculate(row['ExtroScore'], row['ResScore']), axis = 1)
dfnew.head()
```

```python
dfnew['Openness'] = dfnew.apply(lambda row: calculate(row['ConScore'], row['OpenScore']), axis = 1)
dfnew['Conscientiousness'] = dfnew.apply(lambda row: calculate(row['DisScore'], row['DepScore']), axis = 1)
```

```python
#no multicollinearity seems present
#check for homosceedasticity is performed below
```

```python
mr = mr.rename(columns = {'Total Belief score /55': 'Total_Belief'})
print(mr)
print(mr.columns)
print(mr.shape)
```

```python
# Create a multiple regression model
model = smf.ols(formula='Total_Belief ~ Agreeableness + Extraversion + Openness + Conscientiousness', data=mr)

# Fit the model
results = model.fit()

# Print the summary
print(results.summary())
```

```python
#As we can see, personality types are a poor predictor of conspiracy belief (R-square = 0.106). Agreeableness is the only statistically significant variable P<0.05).



#Checking the assumption of homoscedasticity

# Get the residuals and predicted values
residuals = results.resid
predicted_values = results.fittedvalues

# Create a scatter plot of the residuals against the predicted values
plt.scatter(predicted_values, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Scatter Plot of Residuals vs Predicted Values")
plt.show()
```
